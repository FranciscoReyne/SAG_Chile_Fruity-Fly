````python
from pyproj import Transformer

# Definir la transformaci√≥n de coordenadas UTM (Zona 19 Sur) a Lat/Lon
transformer = Transformer.from_crs("EPSG:32719", "EPSG:4326")

# Aplicar la conversi√≥n a todas las filas en las columnas 12 y 13
df_numeros[['lat', 'lon']] = df_numeros.apply(lambda row: transformer.transform(row.iloc[12], row.iloc[13]), axis=1, result_type='expand')

# Mostrar el DataFrame con las coordenadas corregidas
display(df_numeros)
````

# Etapa 2: de pdf a pandas dataframe.

Podemos usar **Tabula** o **Camelot** para extraer tablas de un PDF y convertirlas en un **DataFrame de Pandas**.  

üîπ **Tabula** (`tabula-py`)  
- Ideal para PDFs con tablas bien estructuradas.  
- Requiere **Java** instalado.  
- Instalaci√≥n: `pip install tabula-py`  
- Uso b√°sico:  
  ```python
  from tabula import read_pdf
  df = read_pdf("archivo.pdf", pages="all")
  ```

üîπ **Camelot**  
- Funciona mejor con PDFs que tienen **l√≠neas visibles** en las tablas.  
- Instalaci√≥n: `pip install camelot-py`  
- Uso b√°sico:  
  ```python
  import camelot
  tables = camelot.read_pdf("archivo.pdf", pages="1")
  df = tables[0].df
  ```

Si el PDF tiene tablas complejas, y no es suficiente con alguna de estas dos opciones podr√≠amos probar **pdfplumber** para extraer datos con m√°s control.üöÄ
