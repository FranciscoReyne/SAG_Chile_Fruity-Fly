Transformación de datos gps para ser leido por folium o kepler

```python
import pandas as pd
import folium
from pyproj import Transformer

# Definir el sistema de referencia correcto (ajusta según la zona UTM de Chile)
transformer = Transformer.from_crs("EPSG:5361", "EPSG:4326", always_xy=True)  # UTM 19S a WGS84

# Suponiendo que df_numeros tiene las coordenadas en las columnas 12 y 13
df_numeros[['lat', 'lon']] = df_numeros.apply(lambda row: pd.Series(transformer.transform(row.iloc[12], row.iloc[13])), axis=1)

# Crear mapa centrado en Chile
Map = folium.Map(location=[-33.45, -70.65], zoom_start=6)  # Coordenadas aproximadas de Santiago

# Agregar puntos al mapa
for _, row in df_numeros.iterrows():
    folium.Marker([row['lat'], row['lon']], popup=f"Coordenada: ({row['lat']}, {row['lon']})").add_to(Map)

# Mostrar mapa
Map
```

```python
from pypdf import PdfReader
import camelot
import pandas as pd
from IPython.display import display
from pypdf import PdfWriter
from google.colab import files

# Crear el objeto para combinar PDFs
pdf_writer = PdfWriter()

# Agregar cada PDF al nuevo documento
for pdf in pdf_files.keys():
    pdf_writer.append(pdf)

# Guardar el archivo final
pdf_writer.write("PDF_unido.pdf")
pdf_writer.close()

# Cargar el PDF
pdf_path = "PDF_unido.pdf"
reader = PdfReader(pdf_path)

# Obtener el número de páginas
num_paginas = len(reader.pages)
print(f"El PDF tiene {num_paginas} páginas.")

# Extraer todas las tablas del PDF con una estrategia mejorada
df_list = []  # Lista para almacenar DataFrames de cada página

for page_num in range(1, num_paginas + 1):  # Procesar todas las páginas
    try:
        tables = camelot.read_pdf(pdf_path, pages=str(page_num), flavor="stream")  # Usamos "stream" para mejorar la extracción

        if tables.n > 0:  # Si hay tablas detectadas en la página
            df_page = pd.concat([table.df for table in tables], ignore_index=True)
            df_list.append(df_page)
        else:
            print(f"No se detectaron tablas en la página {page_num}.")

    except Exception as e:
        print(f"Error en la página {page_num}: {e}")

# Unir todos los DataFrames extraídos
if df_list:
    df_total = pd.concat(df_list, ignore_index=True)
else:
    print("No se pudieron extraer tablas del PDF.")
    df_total = pd.DataFrame()  # Creamos un DataFrame vacío en caso de error
display(df_total)
```

```python

# Llenar valores NaN y limpiar espacios vacíos
df_total.replace(r"^\s*$", "NaN", regex=True, inplace=True)  # Reemplazar espacios vacíos con NaN
df_total.fillna("NaN", inplace=True)

# Filtrar filas con coordenadas UTM válidas (columna 12 y 13)
df_limpio = df_total.dropna(subset=[12])  # Eliminar filas donde la columna 12 es NaN

# Convertir valores numéricos correctamente
df_numeros = df_limpio[pd.to_numeric(df_limpio[12], errors='coerce').notna()]

print(f"El DataFrame final tiene {df_numeros.shape[0]} filas.")
display(df_numeros)
```
